## Data Summary

**Date:** The trading date. It uniquely identifies the market day.

**Open:** The stock price at market opening for the given trading day

**High:** The highest price reached by the stock during the trading session
 
**Low:** The lowest price reached by the stock during the trading session.

**Close:** The stock price at market close. 

**Adj Close:** The closing price adjusted to account for corporate actions such as stock splits and dividends. It provides a more accurate view of the stockâ€™s true historical performance and is commonly used for long-term analysis and return calculations.

**Volume:** The total number of shares traded during the day, representing market activity and liquidity.

## 2. Distribution Fitting & Normality Testing

```python
mu_norm, std_norm = stats.norm.fit(returns)
df_t, mu_t, std_t = stats.t.fit(returns)

print(f"Normal fit:    Î¼ = {mu_norm:.6f},  Ïƒ = {std_norm:.6f}")
print(f"Student-t fit: df = {df_t:.2f},  Î¼ = {mu_t:.6f},  Ïƒ = {std_t:.6f}")

# Histogram with both PDFs overlaid
fig, ax = plt.subplots(figsize=(12, 5))
ax.hist(returns, bins=100, density=True, alpha=0.5, color="skyblue", edgecolor="white", label="Empirical")

x = np.linspace(returns.min(), returns.max(), 500)
ax.plot(x, stats.norm.pdf(x, mu_norm, std_norm), "r-", lw=2, label=f"Normal (Î¼={mu_norm:.4f}, Ïƒ={std_norm:.4f})")
ax.plot(x, stats.t.pdf(x, df_t, mu_t, std_t), "g--", lw=2, label=f"Student-t (df={df_t:.1f})")

ax.set_xlabel("Daily Return")
ax.set_ylabel("Density")
ax.set_title("Returns Distribution: Normal vs Student-t Fit")
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# QQ-Plot: visual normality check
fig, ax = plt.subplots(figsize=(6, 6))
stats.probplot(returns, dist="norm", plot=ax)
ax.set_title("QQ-Plot of Daily Returns vs Normal Distribution")
plt.tight_layout()
plt.show()
```

# Normality tests

```python
sample = returns.sample(min(5000, len(returns)), random_state=42)
stat_sw, p_sw = stats.shapiro(sample)

stat_jb, p_jb = stats.jarque_bera(returns)

print("Normality Tests \n")
print(f"Shapiro-Wilk:  statistic = {stat_sw:.6f},  p-value = {p_sw:.2e}")
print(f"Jarque-Bera:   statistic = {stat_jb:.2f},  p-value = {p_jb:.2e}")
print()
if p_sw < 0.05:
    print("Shapiro-Wilk: REJECT Hâ‚€ (returns are NOT normally distributed at Î±=0.05)")
else:
    print("Shapiro-Wilk: FAIL to reject Hâ‚€ (returns may be normally distributed)")
if p_jb < 0.05:
    print("Jarque-Bera:  REJECT Hâ‚€ (returns are NOT normally distributed at Î±=0.05)")
else:
    print("Jarque-Bera:  FAIL to reject Hâ‚€ (returns may be normally distributed)")
```

## 3. Confidence Intervals

Compute the 95% confidence interval for the mean daily return using the t-distribution:

$$\bar{x} \pm t_{\alpha/2,\, n-1} \cdot \frac{s}{\sqrt{n}}$$

Then compute per-year CIs and visualize them.

```python
# 95% Confidence Interval for overall mean daily return
n = len(returns)
mean_r = returns.mean()
se = returns.std(ddof=1) / np.sqrt(n)

ci_low, ci_high = stats.t.interval(confidence=0.95, df=n - 1, loc=mean_r, scale=se)

print("=== 95% Confidence Interval for Mean Daily Return ===\n")
print(f"Sample size:    n = {n}")
print(f"Sample mean:    xÌ„ = {mean_r:.6f}")
print(f"Std error:      SE = {se:.6f}")
print(f"95% CI:         [{ci_low:.6f}, {ci_high:.6f}]")
```

The CI does NOT contain 0: evidence that the mean return is significantly positive.

```python
# Per-year 95% confidence intervals for mean return
yearly_ci = (
    data.dropna(subset=["Returns"])
    .groupby("Year")["Returns"]
    .agg(["mean", "std", "count"])
)
yearly_ci["se"] = yearly_ci["std"] / np.sqrt(yearly_ci["count"])
yearly_ci["ci_low"] = yearly_ci["mean"] - 1.96 * yearly_ci["se"]
yearly_ci["ci_high"] = yearly_ci["mean"] + 1.96 * yearly_ci["se"]

# Plot
fig, ax = plt.subplots(figsize=(14, 5))
years = yearly_ci.index
ax.errorbar(
    years,
    yearly_ci["mean"],
    yerr=1.96 * yearly_ci["se"],
    fmt="o-",
    capsize=4,
    color="steelblue",
    ecolor="coral",
    label="Mean Â± 95% CI"
)
ax.axhline(0, color="black", linestyle="--", linewidth=0.8)
ax.set_xlabel("Year")
ax.set_ylabel("Mean Daily Return")
ax.set_title("95% Confidence Interval for Mean Daily Return by Year")
ax.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 4. Hypothesis Testing

### 4a. One-sample t-test
$H_0$: The mean daily return is zero ($\mu = 0$)  
$H_1$: The mean daily return is not zero ($\mu \neq 0$)

```python
# One-sample t-test: Hâ‚€: Î¼ = 0
t_stat, p_value = stats.ttest_1samp(returns, popmean=0)

print("=== One-Sample t-Test: Is the mean daily return â‰  0? ===\n")
print(f"t-statistic: {t_stat:.4f}")
print(f"p-value:     {p_value:.4e}")
print()
alpha = 0.05
if p_value < alpha:
    print(f"â†’ p < {alpha}: REJECT Hâ‚€. The mean daily return is significantly different from zero.")
else:
    print(f"â†’ p â‰¥ {alpha}: FAIL to reject Hâ‚€. No significant evidence that mean return differs from zero.")
```

### 4b. Two-sample t-test
$H_0$: Mean returns are equal pre-COVID and post-COVID  
$H_1$: Mean returns differ between the two periods  

Split point: March 2020

```python
# Two-sample t-test: pre-COVID vs post-COVID returns
covid_split = pd.Timestamp("2020-03-01")

pre_covid = data.loc[(data["Date"] < covid_split), "Returns"].dropna()
post_covid = data.loc[(data["Date"] >= covid_split), "Returns"].dropna()

t_stat2, p_value2 = stats.ttest_ind(pre_covid, post_covid, equal_var=False)  # Welch's t-test

print("=== Two-Sample t-Test (Welch's): Pre-COVID vs Post-COVID Returns ===\n")
print(f"Pre-COVID:  n = {len(pre_covid)},  mean = {pre_covid.mean():.6f},  std = {pre_covid.std():.6f}")
print(f"Post-COVID: n = {len(post_covid)},  mean = {post_covid.mean():.6f},  std = {post_covid.std():.6f}")
print(f"\nt-statistic: {t_stat2:.4f}")
print(f"p-value:     {p_value2:.4e}")
print()
if p_value2 < 0.05:
    print("â†’ REJECT Hâ‚€: Mean returns significantly differ between the two periods.")
else:
    print("â†’ FAIL to reject Hâ‚€: No significant difference in mean returns.")
```

### 4c. Chi-squared test of independence
$H_0$: The proportion of "up" days is the same across all months  
$H_1$: The proportion of "up" days differs across months

```python
# Chi-squared test: are "up" vs "down" days independent of month?
data_chi = data.dropna(subset=["Returns"]).copy()
data_chi["Direction"] = np.where(data_chi["Returns"] >= 0, "Up", "Down")

contingency = pd.crosstab(data_chi["Month"], data_chi["Direction"])
print("Contingency Table (Month Ã— Direction):\n")
print(contingency)

chi2, p_chi, dof, expected = stats.chi2_contingency(contingency)

print(f"\n=== Chi-Squared Test of Independence ===")
print(f"Ï‡Â² statistic: {chi2:.4f}")
print(f"Degrees of freedom: {dof}")
print(f"p-value: {p_chi:.4e}")
print()
if p_chi < 0.05:
    print("â†’ REJECT Hâ‚€: The proportion of up/down days depends on the month.")
else:
    print("â†’ FAIL to reject Hâ‚€: Up/down day proportions are independent of month.")
```

# Monte Carlo Simulation

We simulate future price paths GBM

The daily step formula is:

$$S_{t+1} = S_t \cdot \exp\left((\mu - \tfrac{\sigma^2}{2}) \cdot \Delta t + \sigma \cdot \sqrt{\Delta t} \cdot Z\right)$$

where $Z \sim \mathcal{N}(0,1)$ is a random draw from the standard normal distribution.

```python
# Starting from the last known closing price
S0 = data["Close"].iloc[-1]

# Daily mean return and standard deviation from historical data
mu = returns.mean()
sigma = returns.std()

# Time horizon: 252 trading days (~1 year)
T = 252

# Number of simulated scenarios
n_sim = 10000

print(f"Initial price (S0):          ${S0:.2f}")
print(f"Mean daily return (mu):      {mu:.6f}")
print(f"Daily volatility (sigma):    {sigma:.6f}")
print(f"Number of simulations:       {n_sim}")
print(f"Horizon:                     {T} trading days (~1 year)")
```

## Random path generation (Box-Muller)

To generate $\mathcal{N}(0,1)$ random numbers we use `np.random.rand()` (uniform distribution) and apply the **Box-Muller transform**:

$$Z = \sqrt{-2 \ln(U_1)} \cdot \cos(2\pi U_2)$$

where $U_1, U_2 \sim \text{Uniform}(0,1)$.

```python
np.random.seed(42)

# Generate two uniform random matrices for Box-Muller
U1 = np.random.rand(T, n_sim)
U2 = np.random.rand(T, n_sim)

# Box-Muller transform: uniform -> standard normal
Z = np.sqrt(-2 * np.log(U1)) * np.cos(2 * np.pi * U2)

# Quick sanity check
print(f"Z mean:  {Z.mean():.4f}  (expected ~0)")
print(f"Z std:   {Z.std():.4f}  (expected ~1)")

# Simulated daily multiplicative returns using GBM
sim_returns = np.exp((mu - 0.5 * sigma**2) + sigma * Z)

# Build price paths starting from S0
price_paths = np.zeros((T + 1, n_sim))
price_paths[0] = S0

for t in range(1, T + 1):
    price_paths[t] = price_paths[t - 1] * sim_returns[t - 1]

print(f"\nGenerated {n_sim} paths over {T} trading days.")
```

## Simulated price paths

Fan chart showing percentile bands across all scenarios.

```python
sns.set_theme(style="ticks")
palette = sns.color_palette("rocket_r", n_colors=200)

days = np.arange(T + 1)

# Pick 200 paths and sort by final price so colors map low -> high
sample_idx = np.random.default_rng(42).choice(n_sim, size=200, replace=False)
final_vals = price_paths[-1, sample_idx]
order = np.argsort(final_vals)
sample_idx = sample_idx[order]

fig, ax = plt.subplots(figsize=(14, 6))

for rank, idx in enumerate(sample_idx):
    ax.plot(days, price_paths[:, idx], color=palette[rank], alpha=0.5, linewidth=0.7)

# Median path
perc_50 = np.percentile(price_paths, 50, axis=1)
ax.plot(days, perc_50, color="white", linewidth=2.5, zorder=5)
ax.plot(days, perc_50, color="black", linewidth=1.5, zorder=6, label="Median")

ax.axhline(S0, color="black", linestyle="--", linewidth=1, label=f"Initial price: ${S0:.2f}")

ax.set_xlabel("Trading Days")
ax.set_ylabel("Simulated Price ($)")
ax.set_title(f"Monte Carlo Simulation â€“ {n_sim} paths over {T} days")
ax.legend()
sns.despine()
plt.tight_layout()
plt.show()
```

## Final price distribution

Distribution of simulated prices at day $T = 252$ (~1 year). As expected from GBM, the distribution is **log-normal**.

```python
sns.set_theme(style="ticks")
palette = sns.color_palette("rocket_r", 5)

# Final prices (last row of the matrix)
final_prices = price_paths[-1]

fig, ax = plt.subplots(figsize=(12, 5))
ax.hist(final_prices, bins=80, density=True, color=palette[2], edgecolor="white", alpha=0.8)

ax.axvline(S0, color="black", linestyle="--", linewidth=1.5, label=f"Initial price: ${S0:.2f}")
ax.axvline(np.median(final_prices), color=palette[4], linewidth=2, label=f"Median: ${np.median(final_prices):.2f}")
ax.axvline(np.mean(final_prices), color=palette[0], linewidth=2, label=f"Mean: ${np.mean(final_prices):.2f}")

ax.set_xlabel("Price ($)")
ax.set_ylabel("Density")
ax.set_title(f"Simulated Price Distribution after {T} Trading Days")
ax.legend()
sns.despine()
plt.tight_layout()
plt.show()

# Mean > Median confirms the right-skewed (log-normal) shape
print(f"Mean > Median: right-skewed distribution (log-normal)")
```

## Summary statistics and Value at Risk (VaR)

**VaR at 95%** tells us: with 95% confidence the price will NOT fall below this value. Only a 5% chance of a worse outcome.

```python
# Descriptive statistics on simulated final prices
print("Final price statistics:")
print(f"  Initial price:    ${S0:.2f}")
print(f"  Mean:             ${np.mean(final_prices):.2f}")
print(f"  Median:           ${np.median(final_prices):.2f}")
print(f"  Std deviation:    ${np.std(final_prices):.2f}")
print(f"  Min:              ${np.min(final_prices):.2f}")
print(f"  Max:              ${np.max(final_prices):.2f}")
print()

# Value at Risk: 5th and 1st percentile of final prices
var_95 = np.percentile(final_prices, 5)
var_99 = np.percentile(final_prices, 1)

print("Value at Risk (VaR):")
print(f"  VaR 95%: ${var_95:.2f}  (5% probability the price falls below)")
print(f"  VaR 99%: ${var_99:.2f}  (1% probability the price falls below)")
print()

# Scenario probabilities using array indexing
scenarios_up     = final_prices[final_prices > S0]
scenarios_down   = final_prices[final_prices < S0]
scenarios_double = final_prices[final_prices > 2 * S0]

print("Estimated probabilities:")
print(f"  P(price goes up):      {scenarios_up.size / n_sim * 100:.1f}%")
print(f"  P(price goes down):    {scenarios_down.size / n_sim * 100:.1f}%")
print(f"  P(price doubles):      {scenarios_double.size / n_sim * 100:.1f}%")
```
